##ステップ関数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#where関数 np.where(条件,trueの時の結果, falseの時の結果)
def step_function(x):
  return np.where(x<=0,0,1)

x = np.linspace(-5,5)
y = step_function(x)

plt.plot(x,y)
plt.show()
```

##シグモイド関数
$$\frac{1}{1+exp(-x)}$$
微分が扱いやすいという特徴がある

```python
def sigmoid_function(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-5,5)
y = sigmoid_function(x)

plt.plot(x,y)
plt.show()
```

##tanh
numpyのtanhを使うことで簡単に実装できる
0で対照になっている

```python
def tanh_function(x):
  return np.tanh(x)

x = np.linspace(-5,5)
y = tanh_function(x)

plt.plot(x,y)
plt.show()
```

##ReLu ランプ関数
出力層以外の活性化関数では使われている

```python
def relu_function(x):
  return np.where(x<=0,0,x)

x = np.linspace(-5,5)
y = relu_function(x)

plt.plot(x,y)
plt.show()
```

##恒等関数
xの関数をそのまま出力する
出力層の活性化関数に用いられる

```python
x = np.linspace(-5,5)
y = x

plt.plot(x,y)
plt.show()
```

##ソフトマックス関数
$$\frac{exp(x)}{\sum_{k=1}^{n}exp(x_k)}$$
全てを足し合わせると1になるので確率を表す
$$\sum_{l=1}^{n}\frac{exp(x_l)}{\sum_{k=1}^{n}exp(x_k)}=1$$



```python
def softmax_function(x):
  return np.exp(x)/np.sum(np.exp(x))

y = softmax_function(np.array([1,2,3]))
print(y)
```
